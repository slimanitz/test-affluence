```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
from dateutil.parser import parse
```

```python
df_record = pd.read_csv('records.csv')
df_sensors = pd.read_csv('sensors_site.csv')
df_sites = pd.read_csv('site_timetables.csv')
```

```python
now = datetime(2023,4,3,14,00)
now.strftime("%c")
before_2_hours = now - timedelta(hours=2)
yesterday =  now - timedelta(hours=24)
yesterday.strftime("%c")
daybeforeyesterday =  now - timedelta(hours=48)
daybeforeyesterday.strftime("%c")

```

```python
df_sensors.head()
```

```python
df_record.head()
df_record.sort_values(by="record_datetime")
```

```python
df_sites.head()
```

```python
#Removing closed sites
df_sites.dropna(subset=["opening_datetime_utc", "closing_datetime_utc"], how='all',inplace=True)
df_sites.sort_values(by="opening_datetime_utc")
```

```python
df_sensors =  df_sensors[df_sensors['site_id'].isin(df_sites["site_id"])]
df_sensors
```

```python
df_record = df_record[df_record['sensor_id'].isin(df_sensors['sensor_id'])]
df_record
```

```python
df_record['record_datetime']= pd.to_datetime(df_record['record_datetime'])
df_record
```

```python
df_record= df_record[df_record["record_datetime"] <= now]
df_record
```

```python
df_not_active = df_record[(df_record['entries'] == 0) & (df_record['exits'] == 0)]    
```

```python
df_not_active
```

```python
df_sensor_time = df_sensors.merge(df_sites,how="left",on="site_id")
df_sensor_time
```

```python
df_result = df_sensor_time.merge(df_not_active,how="left",on="sensor_id")
df_result
```

```python
df_result['opening_datetime_utc']= pd.to_datetime(df_result['opening_datetime_utc'])
df_result['closing_datetime_utc']= pd.to_datetime(df_result['closing_datetime_utc'])
```

```python
df_result = df_result[ (df_result['record_datetime'] <= df_result['closing_datetime_utc'])]
```

```python
df_result.info()
```

```python
submit = df_result.groupby(['sensor_id','site_id','sensor_name'])['record_datetime'].min()
```

```python
submit
```

```python
not_working = submit.reset_index()
```

```python
alert3 = not_working[(not_working['record_datetime'] <= daybeforeyesterday) ]
alert2 = not_working[(not_working['record_datetime'] > daybeforeyesterday) & (not_working['record_datetime'] <= yesterday) ]
alert1 = not_working[(not_working['record_datetime'] > yesterday) & (not_working['record_datetime'] <= before_2_hours) ]

alert3
```

```python
alert3['alert_level'] = 3
alert2['alert_level'] = 2
alert1['alert_level'] = 1
```

```python
alert3['last_record_datetime']=alert3['record_datetime']
alert2['last_record_datetime']=alert2['record_datetime']
alert1['last_record_datetime']=alert1['record_datetime']
```

```python
result = pd.concat([alert3, alert1,alert2], axis=0)
result
```

```python
result['alert_datetime'] = now
result.drop(['record_datetime'],axis=1,inplace=True)
```

# Inserting into the database

```python
from sqlalchemy import create_engine
from sqlalchemy.orm import declarative_base
from sqlalchemy import Column, Integer, String, DateTime
from sqlalchemy.dialects.mysql import ENUM
import MySQLdb
```

```python
from enum import IntEnum
class AlertSeverity(IntEnum):
    ONE = 1
    TWO = 2
    THREE=3
```

```python
Base = declarative_base()
class SensorsAlert(Base):
     __tablename__ = "sensors_alerts"

     alert_id = Column(Integer, primary_key=True)
     sensor_name = Column(String(45))
     sensor_id = Column(String(45))
     alert_level  = Column(ENUM(AlertSeverity))
     alert_datetime = Column(DateTime)
     site_id = Column(Integer)
     last_record_datetime = Column(DateTime)
```

```python
database_username = 'ecom'
database_password = 'password'
database_ip       = '127.0.0.1:3306'
database_name     = 'ecom'
database_connection = create_engine('mysql://{0}:{1}@{2}/{3}'.
                                               format(database_username, database_password, 
                                                      database_ip, database_name))
Base.metadata.create_all(bind=database_connection)
```

```python
result.reset_index(drop=True,inplace=True)
```
